```python
import pandas
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def plot(file):
    
    df = pandas.read_csv(file)
    
    #convert timestamp
    df['timestamp'] = pandas.to_datetime(df['Data'], format = '%Y-%m-%d')
    
    df.set_index('timestamp', inplace = True)
    df.index.to_pydatetime() 
    
    # plot dataframe
    ax = df.plot(figsize=(20,10), lw=2, 
                 colormap='jet', title='Dieta', 
                 grid="True")
                 #yticks=([83.5,87.7, 92.4, 93.4]))
    
    ax.set_xlabel("Data")
    ax.set_ylabel("Peso (Kg)")
    
    
    # show dataframe as table
    display(df)
```

```python
plot('/home/alberto/Downloads/dieta.csv')
```
